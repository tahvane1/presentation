```python
def calculate_values(applications, onboarding,aiv,offer_rate,acc_rate,lgd,default_rate,fp,tp,fn):
    
    current_cost = round(onboarding*applications,0)
    future_cost  =  round(applications*onboarding*(tp*(offer_rate/100)+
                         (1-offer_rate/100)*fp) ,0)

    rejection_loss = round(applications*(offer_rate/100)*(acc_rate/100)*fn*aiv*((1-default_rate/100)-default_rate/100*lgd/100)) #B1*B2*B9*B4*B5
    total_boarding = future_cost + rejection_loss
    total_savings  =  current_cost - total_boarding  
    return [current_cost,future_cost,rejection_loss, total_boarding,total_savings]
    
    
```

### Equations used
$
\text{Total onboarding now} = \text{applications} \times \text{onboarding cost}
$
<br/>
$
\text{Onboarding with filtering} = \text{applications} \times  \text{onboarding cost} \times  
$
$$
\left( \frac{\text{offer rate}}{100}  \times  \text{true positive rate} + \left( 1 - \frac{\text{offer rate}}{100}   \right) \times \text{false positive rate} \right)
$$
<br/>
$
\text{Customer rejection loss} = \text{applications} \times \frac{\text{offer rate}}{100}     \times \frac{\text{offer acceptance rate}}{100} \times \text{false negative rate}  \times \text{loan value} \times
$
$$\left(\left( 1 - \frac{\text{default rate}}{100} \right)   - \frac{\text{default rate}}{100} \times \frac{\text{lgd}}{100} \right)
$$
<br/>
$
\text{Total costs with filtering} = \text{Onboarding with filtering} + \text{Customer rejection loss}
$
<br/>
$
\text{Potential savings before MANU cost} = \text{Total onboarding now} - \text{Total costs with filtering}
$
